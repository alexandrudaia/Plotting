```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
#Reading the correspoding data
df=pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
```

```python
df.columns
```

```python
#splot   Date ito   year  and  minth  date  day date
df['Year'], df['Month-Date'] = zip(*df['Date'].apply(lambda x: (x[:4], x[5:])))
#remove    29 febryary in order   to have nice loocking   plots
df = df[df['Month-Date'] != '02-29']
```

```python
 
df.head()
```

```python
#aggregated  min and max  temperature   grouped  my month  date
import numpy as np
min_temp = df[(df['Element'] == 'TMIN') & (df['Year'] != '2015')].groupby('Month-Date').aggregate({'Data_Value':np.min})
max_temp = df[(df['Element'] == 'TMAX') & (df['Year'] != '2015')].groupby('Month-Date').aggregate({'Data_Value':np.max})
```

```python
max_temp.head()
```

```python
min_temp.head()
```

```python
#determine minimum and  maximum for  2015 for each month
minimum2015= df[(df['Element'] == 'TMIN') & (df['Year'] == '2015')].groupby('Month-Date').aggregate({'Data_Value':np.min})
maximum2015 = df[(df['Element'] == 'TMAX') & (df['Year'] == '2015')].groupby('Month-Date').aggregate({'Data_Value':np.max})
```

```python
minimum2015.head()
```

```python
maximum2015.head()
```

```python
##################Determining  the  broken up points###########################
brokedMin = np.where(minimum2015['Data_Value'] < min_temp['Data_Value'])[0]
brokedMax = np.where(maximum2015['Data_Value'] > max_temp['Data_Value'])[0]
print(brokedMin)
print(brokedMax)
```

```python
#################Macking the pretty plot###############################################
%matplotlib inline
import  matplotlib  as  matplotlib
import  matplotlib.pyplot as plt
plt.figure()
plt.figure = matplotlib.pyplot.gcf()
plt.figure.set_size_inches(18.5, 10.5)
plt.plot(min_temp.values, 'b', label = 'Low recodring')
plt.plot(max_temp.values, 'r', label = 'Max recording')
plt.scatter(brokedMin ,minimum2015.iloc[brokedMin], s = 20, c = 'g', label = 'Broked Low')
plt.scatter(brokedMax , maximum2015.iloc[brokedMax], s = 20, c = 'm', label = 'Broked High')
plt.gca().axis([-10, 400, -150, 700])
plt.xticks(range(0, len(min_temp), 20), min_temp.index[range(0, len(min_temp), 20)], rotation = '45')
plt.xlabel('Day ')
plt.ylabel('Temperature ')
plt.title('Temperature Detroit')
plt.legend(loc = 1, frameon = False)
plt.gca().fill_between(range(len(min_temp)),min_temp['Data_Value'],max_temp['Data_Value'], facecolor = 'yellow', alpha = 0.5)
 
plt.show()
```
